```python
df = df[df['State'] != 'Total (All India)']
df = df[df['State'] != 'Total (Uts)']
df = df[df['State'] != 'Total (States)']
```

```python
yearwise_total_suicide = df.groupby("Year")["Total"].sum()
print('Total number of suicides in each year: ', yearwise_total_suicide)

#plotting this in a line graph
yearwise_total_suicide.plot(kind="line", figsize = (9,5), color = 'r')
plt.title('Yearwise Total Suicides')
plt.xlabel('Year')
plt.ylabel('Number of suicides')
plt.show()
```

This trend indicates that the issue of suicides has been growing over the years, reaching a peak in 2011, and then slightly decreasing in 2012. 

```python
statewise_total_suicide = df.groupby("State")["Total"].sum().sort_values(ascending=False)
print('Total number of suicides in each state: ', statewise_total_suicide)

# Plotting this in a bar graph
statewise_total_suicide.plot(kind="bar", figsize=(15,7), color='r')
plt.title('Total Suicides by State')
plt.xlabel('State')
plt.ylabel('Number of Suicides')
plt.show()
```

```python
# Grouping the data by Year and Type and summing the Total
causes_yearwise_suicide = df.groupby(['Year', 'Type_code'])['Total'].sum().unstack()

# Plotting the data
causes_yearwise_suicide.plot(kind='bar', stacked=True, figsize=(15, 10))
plt.title('Suicides According To The Causes In Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
#Filtering the data according to 'type_code','Total' and 'Year'
filter_type_code = pd.DataFrame(df.groupby(["Type_code","Year"])["Total"].sum()).reset_index()

#Plotting the above data
sns.catplot(x="Type_code", y="Total", hue="Year", kind="bar", data=filter_type_code, height=7.5, aspect=10/7.5)
plt.title('Suicide According To Causes In Each Year')
```

```python
# Grouping the data by Gender and summing the Total
genderwise_suicide = df.groupby('Gender')['Total'].sum().reset_index()

# Plotting the data
plt.figure(figsize=(8, 8))
plt.pie(genderwise_suicide['Total'], labels=genderwise_suicide['Gender'], autopct='%1.1f%%', startangle=140)
plt.title('Suicide According To Gender')
plt.show()
```

```python
education_gender_suicide = df[df['Type_code'] == 'Education_Status'].groupby(['Type', 'Gender'])['Total'].sum().unstack()

education_gender_suicide.plot(kind='bar', figsize=(12, 8))
plt.title('Suicides by Educational Background (Male vs Female)')
plt.xlabel('Educational Background')
plt.ylabel('Number of Suicides')
plt.legend(title='Gender')
plt.show()
```

- The highest number of suicides for both males and females is among individuals with "No Education" and "Primary" education levels.
- The number of suicides decreases as the education level increases, with the lowest numbers observed among those with "Post Graduate and Above" education.
- Males have a higher number of suicides compared to females across all educational backgrounds.
